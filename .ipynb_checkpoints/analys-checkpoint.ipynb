```python
import pandas as pd
import re
```

```python
data = pd.read_csv('comments.csv')
data.drop('email',1, inplace =True)
```

```python
data.head()
```

```python
rating_summary = data.course_rating.value_counts().to_frame()
```

```python
rating_summary.reset_index(inplace=True)
```

```python
rating_summary
```

```python
data.course_rating = data.course_rating.map(lambda rate: int(re.search('امتیاز (.*) از', rate).group(1)) ,na_action ='ignore')
```

```python
data.head()
```

```python
data.columns
```

```python
data.drop('Unnamed: 0', 1, inplace=True)
```

```python
data.head()
```

```python
data.price = data.price.map(lambda price: 0 if price =='رایگان!' else price)
```

```python
data.head(50)
```

```python
def price_mapper(price):
    dic = { 
    '۰':0, 
    '١':1, 
    '۱':1,
    '٢':2, 
    '۲':2,
    '۳':3, 
    '۴':4, 
    '۵':5, 
    '۶':6, 
    '۷':7, 
    '۸':8, 
    '۹':9, 
    ',':''
    }
    str_price = re.search('(.*)\xa0تومان',price ).group(1)
    return int(''.join([str(dic[x]) for x in str_price]))
```

```python
data.price = data.price.map(lambda x: price_mapper(x) if x!=0 else 0)
```
